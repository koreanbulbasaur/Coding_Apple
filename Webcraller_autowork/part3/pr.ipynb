```python
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_table('./data/income.txt')
df
```

```python
df = df.dropna()

plt.scatter(df['age'], df['income'])
plt.show()
```

```python
def fun(x, a, b, c):
    return a*x + b*x**2 + c
```

```python
opt, cov = curve_fit(fun, df['age'], df['income'])
opt
```

```python
a, b, c = opt
```

```python
import numpy as np
```

```python
x = np.array(range(1, 7))
plt.scatter(df['age'], df['income'])
plt.plot(x, fun(x, a, b, c))
plt.show()
```

```python
import statsmodels.api as st

x = np.column_stack([df['age'], df['age']**2, np.ones(79)])
model = st.OLS(df['income'], x).fit()
model.summary()
```
